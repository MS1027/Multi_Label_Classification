```python
# 라이브러리 불러오기

from selenium import webdriver
from pandas.io.html import read_html
import time
import pandas as pd
import numpy as np
```

## 함수 정의

```python
def defensive (url, sleep_time) :
    # 웹 드라이버 열기 
    chrome_driver = 'chromedriver'
    driver = webdriver.Chrome(chrome_driver)
    driver.get(url)
    
    """
    # 처음 접속 시, 쿠키 동의
    time.sleep(sleep_time)
    accept = driver.find_element('css selector', '#qcCmpButtons > button:nth-child(2)')
    accept.click()
    """
    
    # defensive 버튼 클릭
    time.sleep(sleep_time)
    defense = driver.find_element('link text', 'Defensive')
    defense.click()
    
    # 총 페이지 수 가져오기
    time.sleep(sleep_time)
    page = driver.find_element('link text', 'last')
    total_page = int(page.get_attribute('data-page'))
    
    # 데이터프레임 생성
    df_defensive = pd.DataFrame(columns = ['Player', 'Player.1', 'Apps', 'Mins', 'Tackles', 'Inter', 'Fouls',
       'Offsides', 'Clear', 'Drb', 'Blocks', 'OwnG', 'Rating'])
    
    # 표 스크래핑
    for i in np.arange(total_page)+1 :
        time.sleep(sleep_time)
        table = driver.find_element('xpath', '//*[@id="statistics-table-defensive"]')
        table_html= table.get_attribute('innerHTML')
        df2 = read_html(table_html)[0]
        df_defensive = pd.concat([df_defensive, df2], axis=0)
        driver.find_element('link text','next').click()
        
    return(df_defensive)
```

```python
def offensive (url, sleep_time) :
    # 웹 드라이버 열기 
    chrome_driver = 'chromedriver'
    driver = webdriver.Chrome(chrome_driver)
    driver.get(url)
    
    """
    # 처음 접속 시, 쿠키 동의
    time.sleep(sleep_time)
    accept = driver.find_element('css selector', '#qcCmpButtons > button:nth-child(2)')
    accept.click()
    """
    
    # offensive 버튼 클릭
    time.sleep(sleep_time)
    offense = driver.find_element('link text', 'Offensive')
    offense.click()
    
    # 총 페이지 수 가져오기
    time.sleep(sleep_time)
    page = driver.find_element('link text', 'last')
    total_page = int(page.get_attribute('data-page'))
    
    # 데이터프레임 생성
    df_offensive = pd.DataFrame(columns = ['Player', 'Player.1', 'Apps', 'Mins', 'Goals', 'Assists', 'SpG',
       'KeyP', 'Drb', 'Fouled', 'Off', 'Disp', 'UnsTch', 'Rating'])
    
    # 표 스크래핑
    for i in np.arange(total_page)+1 :
        time.sleep(sleep_time)
        table = driver.find_element('xpath', '//*[@id="statistics-table-offensive"]')
        table_html= table.get_attribute('innerHTML')
        df2 = read_html(table_html)[0]
        df_offensive = pd.concat([df_offensive, df2], axis=0)
        driver.find_element('link text','next').click()
        
    return(df_offensive)
```

```python
def passing (url, sleep_time) :
    # 웹 드라이버 열기 
    chrome_driver = 'chromedriver'
    driver = webdriver.Chrome(chrome_driver)
    driver.get(url)
    
    """
    # 처음 접속 시, 쿠키 동의
    time.sleep(sleep_time)
    accept = driver.find_element('css selector', '#qcCmpButtons > button:nth-child(2)')
    accept.click()
    """
    
    # defensive 버튼 클릭
    time.sleep(sleep_time)
    passing = driver.find_element('link text', 'Passing')
    passing.click()
    
    # 총 페이지 수 가져오기
    time.sleep(sleep_time)
    page = driver.find_element('link text', 'last')
    total_page = int(page.get_attribute('data-page'))
    
    # 데이터프레임 생성
    df_passing = pd.DataFrame(columns = ['Player', 'Player.1', 'Assists', 'KeyP', 'AvgP', 'PS%', 'Crosses',
       'LongB', 'ThrB', 'Rating'])
    
    # 표 스크래핑
    for i in np.arange(total_page)+1 :
        time.sleep(sleep_time)
        table = driver.find_element('xpath', '//*[@id="statistics-table-passing"]')
        table_html= table.get_attribute('innerHTML')
        df2 = read_html(table_html)[0]
        df_passing = pd.concat([df_passing, df2], axis=0)
        driver.find_element('link text','next').click()
        
    return(df_passing)
```

# PL

## defensive

```python
url = 'https://1xbet.whoscored.com/Regions/252/Tournaments/2/Seasons/9075/Stages/20934/PlayerStatistics/England-Premier-League-2022-2023'
```

```python
df_pl_defensive = defensive(url, 3)
df_pl_defensive.head()
```

```python
df_pl_defensive.to_csv('PL_Defensive.csv', index=False)
```

```python
df_pl_offensive = offensive(url, 3)
df_pl_offensive.head()
```

```python
df_pl_offensive.to_csv('PL_Offensive.csv', index=False)
```

## passing

```python
df_pl_passing = passing(url, 3)
df_pl_passing.head()
```

```python
df_pl_passing.to_csv('PL_Passing.csv', index=False)
```
