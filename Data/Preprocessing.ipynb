```python
import pandas as pd
import numpy as np
```

# Function Declaration

```python
def Func(league):
    
    # csv load
    
    # Defensive
    df1 = pd.read_csv('Selenium/' + league + '_Defensive.csv')
    df1.drop(columns=['Player', 'Apps', 'Mins', 'Fouls', 'Offsides', 'Drb', 'OwnG', 'Rating'], inplace=True)

    # Offensive
    df2 = pd.read_csv('Selenium/' + league + '_Offensive.csv')
    df2.drop(columns=['Player', 'Apps', 'Mins', 'Goals', 'Assists', 'Off', 'Disp', 'UnsTch', 'Rating'], inplace=True)


    # Passing
    df3 = pd.read_csv('Selenium/' + league + '_Passing.csv')
    df3.drop(columns=['Player', 'Assists', 'KeyP', 'AvgP', 'Rating', 'Apps', 'Mins'], inplace=True)

    
    # Merge

    df4 = pd.merge(df1, df2)
    df4 = pd.merge(df4, df3)

    
    # Split(Name, Position)

    df4.columns = ['Player', 'Tackles', 'Inter', 'Clear', 'Blocks', 'SpG', 'KeyP', 'Drb', 'Fouled', 'PS%', 'Crosses', 'LongB', 'ThrB']
    df_player = df4[['Player']]
    df_player.astype('str')
    
    df_temp = df_player['Player'].str.split(',', n=2)
    df_name = pd.DataFrame(df_temp.str.get(0))
    df_name.columns = ['Name']
    df_position = pd.DataFrame(df_temp.str.get(2))
    df_position.columns = ['Position']

    df4['Player'] = df_name
    df4['Position'] = df_position

    
    # Drop 'GK' rows
    
    df4.drop(df4[df4['Position'] == ' GK'].index, inplace=True)
    df4 = df4.reset_index(drop=True)

    df_pos = pd.DataFrame([[0 for i in range(8)] for i in range(df4.shape[0])], columns=['CF', 'WF', 'AM', 'WM', 'CM', 'DM', 'FB', 'CB'])

    dic_pos = {'AM(C)':'AM', 'AM(L)':'WF', 'AM(R)':'WF', 'AM(CL)':['WF','AM'], 'AM(CR)':['WF', 'AM'], 'AM(LR)':'WF', 'AM(CLR)':['WF','AM'],
              'M(C)':'CM', 'M(L)':'WM', 'M(R)':'WM', 'M(CL)':['WM','CM'], 'M(CR)':['WM', 'CM'], 'M(LR)':'WM', 'M(CLR)':['WM','CM'],
              'D(C)':'CB', 'D(L)':'FB', 'D(R)':'FB', 'D(CL)':['FB','CB'], 'D(CR)':['FB', 'CB'], 'D(LR)':'FB', 'D(CLR)':['FB','CB'],
              'FW':'CF', 'Forward':'CF', 'Midfielder':'CM', 'DMC':'DM'}

    
    # position rename & labeling
    
    for i in range(df4.shape[0]):
        str_temp = df4.iloc[i, 13]
        temp1 = str_temp.lstrip().split(',')
        temp2 = []
        for pos in temp1:
            temp2.append(dic_pos.get(pos))
        pos_list = []
        for sub in temp2:
            if(type(sub)==list):
                for pos in sub:
                    pos_list.append(pos)
            else:
                pos_list.append(sub)
        for pos in pos_list:
            df_pos.loc[i, pos] = 1

    df_complete = df4.join(df_pos)
    df_complete.drop(columns=['Position'], inplace=True)
    
    return df_complete
```

# Function Call

```python
PL = Func('PL')
LaLiga = Func('LaLiga')
SerieA = Func('SerieA')
Bundes = Func('Bundes')
Ligue1 = Func('Ligue1')
```
